### 2. Pre-processsament

Degut a la natura del senyal electro-encefalogràfic, cal descomposar el senyal en tres bandes de freqüència que el composen: freqüències baixes, mitges i altes. Dit d'altra manera, els senyals EEG es descomposen en la Banda Alpha (α: entre 8 -15 Hz), Banda Beta (β: entre 15-32 Hz) i Banda Gamma (γ: ≥ 32 Hz).

Com que es classifiquen de manera independent, haurem d'executar la nostra classificació tres vegades, una vegada per cada banda de freqüència.

```python
ts2 = np.zeros([n_motiv,n_bands,T,N])
filtered_ts = np.zeros([n_motiv,n_bands,n_trials,T,N])
for i_band in range(n_bands):

    # select band
    freq_band = freq_bands[i_band]

    # band-pass filtering (alpha, beta, gamma)
    n_order = 3
    sampling_freq = 500. # sampling rate

    if freq_band=='alpha':
        low_f = 8./sampling_freq
        high_f = 15./sampling_freq
    elif freq_band=='beta':    
        # beta
        low_f = 15./sampling_freq
        high_f = 32./sampling_freq
    elif freq_band=='gamma':
        # gamma
        low_f = 32./sampling_freq
        high_f = 80./sampling_freq
    else:
        raise NameError('unknown filter')

    # apply filter ts[n_motiv,n_trials,T,N]
    b,a = spsg.iirfilter(n_order, [low_f,high_f], btype='bandpass', ftype='butter')
    filtered_ts[:,i_band,:,:,:] = spsg.filtfilt(b, a, ts, axis=2)

filtered_ts.shape
```

#### 3. Classificació

En aquesta part de la pràctica, classificarem els senyals filtrats. Farem tres classificacions, una per cada banda de freqüència i en compararem el resultat.

Com en qualsevol problema típic de classificació en aprenentatge automàtic, el 80% de les dades són per entrenar el  classificador, i el 20% restant per avaluar la performance de la seva classificació.

**K-Nearest Neighbour**

Amb una imatge d'entrada a classificar, busca les K imatges d'entrenament que més s'hi assemblen per, posteriorment, seleccionar la classe més abundant entre les K possibilitats. En reconeixement de patrons, l'algorisme del k-NN és un mètode per classificar objectes basant-se en els exemples més propers de l'espai de característiques. Aquest classificador treballa cercant primer quines de les K observacions classificades a l'entrenament s'assemblen més a la nova observació a classificar per seguidament assignar-li la classe més comuna entre aquestes K a la nova mostra.

Fem servir l'algoritme k-NN com a mètode de classificació d'estats motivacionals.

**Elecció de la K**

Per fer una bona elecció del paràmetre K ens haurem de basar principalment en les dades. Un cas especial el trobem per K=1 anomenat Nearest Neighbor Algorithm, que es demana de fer servir en aquest cas. 

**Implementació**

Els exemples d'entrenament són vectors en un espai de característiques multidimensional, cada etiqueta té un vector de característiques associat.

 $x i = ( x_{1i} , x_{2i} , . . . , x_{pi}) ∈ X$ 

L'espai és particionat a regions per localitzacions i etiquetes dels exemples d'entrenament. Un punt a l'espai és assignat a la classe C si aquesta és la classe més freqüent entre els k exemples d'entrenament més proper.

La fase de entrenament per k-NN consisteix simplement en emmagatzemar tots els casos coneguts i les seves etiquetes de classe. Si tenim la mostra a classificar (‘I’) i també tenim les noves instàncies (‘t’) hem de seguir els següents passos. Generalment la distància que s'utilitza és la distància euclidiana:

 $d ( x_i , x_j ) = \sqrt( \sum_{r=1}^p (x_i r - x_j r)^2 )$ 

Els exemples d'entrenament són vectors en un espai de característiques multidimensionals, cadascun amb una etiqueta de classe. La fase d'entrenament de l'algorisme consisteix crear models amb elements que tenen assignats les seves característiques. De manera que un cop entrenat, el sistema disposi d'exemples a comparar amb el nou element.

En la fase de classificació, K és una constant definida per l'usuari, i un vector sense marcar (una consulta o punt de prova) es classifica mitjançant l'assignació de l'etiqueta, que és més freqüent entre les K mostres d'entrenament més properes a aquest punt de la consulta. Una mètrica de distància utilitzada per a les variables contínues és la distància euclidiana. Sovint, la precisió de la classificació de k-NN es pot millorar significativament si la distància mètrica s'aprèn amb algorismes especialitzats com a ampli marge les Veí més proper o l'anàlisi dels components de veïnatge.

4. Implementa un classificador 1-NN per classificar les teves dades electro-encefalogràfiques, d'acord amb el template següent:

```python
# classificador 1-NN
class classifier1NN:
    def fit(self, xtrain, ytrain):
        """
        Metode per entrenar el classificador.
        
        :param xtrain: Matriu que conte els vectors de caracteristiques de les dades d'entrenament.
                       A la fila 'i' hi ha el vector de caracteristiques de l'i-essima dada d'entrenament.
        :param ytrain: Vector que conte les etiquetes (en aquest cas l'estat motivacional) de les dades
                       d'entrenament. A la posicio 'i' hi ha l'etiqueta (estat motivacional) de l'i-essima
                       dada d'entrenament.
        """
        self.Xtrain = xtrain
        self.ytrain = ytrain
        
    def predict(self, X):
        """
        Metode que prediu l'etiqueta (en aquest cas estat motivacional) del vector de caracteristiques X.
        
        :param X: Vector de caracteristiques d'una dada de test.
        :return: La prediccio de l'etiqueta (en aquest cas estat motivacional) de la dada que te com a vector
                 de caracteristiques el vector passat per parametre.
        """
        dist = np.linalg.norm(X - self.Xtrain, axis=1) # Calculem la distancia euclidiana entre el vector de
                                                       # de caracteristiques X i cada vector de caracteristiques
                                                       # de les dades d'entrenament
        i = np.argmin(dist) # Prenem l'index del vector de caracteristiques d'entrenament mes proper a X
        # Retornem l'etiqueta (estat motivacional) que correspon a la posicio on hem trobat la distancia minima
        return self.ytrain[i]
```

5. Fes servir la teva classe per classificar les dades per cada banda de freqüència. La teva matriu d'entrada és la matriu filtered_ts. N'haureu d'extreure les dades per cada banda de freqüència i distribuir les dades en dos subconjunts: un 80% de training, i un 20% de testing.

La classificació cal repetir-la per cada banda de freqüència un número N de vegades (N=20 típicament) per poder obtenir una distribució de success rate per cada problema de classificació.

```python
def split_train_test(x, y, test_size=0.2):
    """
    Funcio que, donada una matriu x que conté unes dades i una matriu y que conté les etiquetes de
    les dades, separa el conjunt x en un (1-test_size)*100% de dades de train i un test_size*100% de dades de test.
    
    :param x: matriu 3D de mida AxBxC, on A és el nombre d'etiquetes diferents que tenen les dades (en aquest
               cas A=3 i es el nombre d'estats motivacionals), i fixat A la matriu x[a,:,:] conte els vectors
               de caracteristiques de totes les dades de l'estat motivacional a.
    :param y: matriu 3D de mida AxB, on a la posicio y[a,b] hi ha l'etiqueta (en aquest cas estat motivacional)
               de la dada que te el vector de caracteristiques x[a,b,:]
    :param test_size: percentatge de dades a utiltizar com a test. Ha de ser un numero entre 0 i 1. Per defecte
                      el seu valor es 0.2.
    :return: X_train, X_test, y_train, y_test on:
                - X_train es una matriu 2D amb els vectors de caracteristiques de les dades d'entrenament.
                - X_test es una matriu 2D amb els vectors de caracteristiques de les dades de test.
                - y_train es un vector amb les etiquetes (estat motivacional) de les dades d'entrenament.
                - y_test es un vector amb les etiquetes (estat motivacional) de les dades de test.
    """
    num_train_samples = round(x.shape[1]*(1-test_size)) # Calculem el nombre de dades que ha de tenir el subconjunt de train
    num_test_samples = x.shape[1] - num_train_samples # Calculem el nombre de dades que ha de tenir el subconjunt de test
    
    # Creem un numpy array boolea amb tants True com el nombre de dades que ha de tenir el subconjunt de train
    # i tants False com el nombre de dades que ha de tenir el subconjunt de test
    selected_indexes = np.concatenate((np.full(num_train_samples, True), np.full(num_test_samples, False)))
    # Reordenem els booleans aleatoriament. L'array selected_indexes indicara els indexos de les dades que
    # s'utilitzaran per a train (que seran els indexos on el seu valor sigui True)
    np.random.shuffle(selected_indexes)
    
    # Considerem el complementari de l'array selected_indexes. Aquest array indicara els indexos de les dades que
    # s'utilitzaran per a test (que seran els indexos on el seu valor sigui True)
    not_selected_indexes = np.logical_not(selected_indexes)                                                        
    
    # Seleccionem les dades de train i les transformem en una matriu 2D de dimensio A*num_train_samples x C
    # (en aquest cas 3*num_train_samples x 48)
    X_train = x[:,selected_indexes,:].reshape(x.shape[0]*num_train_samples, x.shape[2])
    # Seleccionem les etiquetes de train i les transformem en un vector de dimensio A*num_train_samples
    # (en aquest cas 3*num_train_samples)
    y_train = y[:, selected_indexes].reshape(y.shape[0]*num_train_samples)
    # Seleccionem les dades de test i les transformem en una matriu 2D de dimensio A*num_test_samples x C
    # (en aquest cas 3*num_test_samples x 48)
    X_test = x[:,not_selected_indexes,:].reshape(x.shape[0]*num_test_samples, x.shape[2])
    # Seleccionem les etiquetes de test i les transformem en un vector de dimensio A*num_test_samples
    # (en aquest cas 3*num_test_samples)
    y_test = y[:, not_selected_indexes].reshape(y.shape[0]*num_test_samples)
    
    return X_train, X_test, y_train, y_test
```

```python
# Classificació
n_reps = 20

n_samples_per_freq = filtered_ts.shape[2] # Prenem el nombre de dades que tenim per a cada banda de frequencia
# Creem una matriu que contingui les etiquetes de les dades, que es corresponen amb l'estat motivacional de les dades
labels = np.array((np.zeros(n_samples_per_freq, dtype=int), np.ones(n_samples_per_freq, dtype=int), np.full(n_samples_per_freq, 2, dtype=int)))
# Matriu que contindra l'accuracy de les classificacions fetes per a cada banda de frequencia i per a cada repeticio
accuracies = np.zeros((n_bands, n_reps))

classifier = classifier1NN() # Instanciem el classificador 1-NN

for n in range(n_bands): # Repetim la classificacio per a cada banda de frequencia
    # Prenem les dades de la banda de frequencia corresponent. Fem el valor absolut de les mostres, ja que realment ens
    # interessa el quadrat de les mostres (i com que estem classificant ho podem simplificar prenent el valor absolut), i 
    # fem la mitjana respecte l'eix temporal per disminuir la quantitat de dades
    samples = np.abs(filtered_ts[:,n,:,:,:]).mean(axis=2) 
    
    for i in range(n_reps): # Fem la classificacio n_reps vegades per poder obtenir un valor mes fiable de l'accuracy
                            # dels estats motivacionals predits
        X_train, X_test, y_train, y_test = split_train_test(samples, labels) # Obtenim les dades de train i de test
        classifier.fit(X_train, y_train) # Entrenem el classificador amb les dades de train
        
        results = list(map(classifier.predict, X_test)) # Fem la prediccio de l'estat motivacional de cada dada de test

        accuracies[n,i] = (np.array(results) == y_test).sum() / y_test.shape[0] # Guardem l'accuracy de la classificacio

accuracies_mean = accuracies.mean(axis=1) # Calculem l'accuracy mitjana per a cada banda de frequencia
accuracies_std = accuracies.std(axis=1) # Calculem la desviacio estandard de l'accuracy per a cada banda de frequencia

# Mostrem l'accuracy mitjana per a cada banda de frequencia
for n in range (n_bands):
    freq = freq_bands[n]
    print("Accuracy using the frequency band {}: {}".format(freq, accuracies_mean[n]))
```
